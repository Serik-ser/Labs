## 5. Прапор Джибуті




```python
def bubble_sort_with_swaps(numbers):
  """
  Сортує список методом бульбашки та повертає кількість переставлень.
  """
  n = len(numbers)
  swaps = 0  # Лічильник переставлень

  # Зовнішній цикл для проходів
  for i in range(n - 1):
    # Внутрішній цикл для порівнянь
    for j in range(0, n - i - 1):
      # Порівнюємо два сусідні елементи
      if numbers[j] > numbers[j + 1]:
        # Якщо вони не по порядку, міняємо їх місцями
        numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
        swaps += 1  # Збільшуємо лічильник переставлень

  return swaps

def task4():
  """
  Головна функція для Завдання 4.
  Отримує ввід, обробляє помилки та викликає сортування.
  """
  print("Введіть список чисел через пробіл:")
  user_input = input()

  try:
    # 1. Розбиваємо рядок на список рядків по пробілу
    str_list = user_input.split()

    # 2. Конвертуємо кожен елемент у ціле число (int)
    #    Це місце, де виникне помилка, якщо введено не число
    numbers = [int(num) for num in str_list]

    # 3. Якщо конвертація успішна, викликаємо сортування
    swap_count = bubble_sort_with_swaps(numbers)

    # 4. Виводимо результат у форматі, як у завданні
    print(f"Було виконано {swap_count} переставлень")

  except ValueError:
    # 5. Обробка помилки, якщо `int(num)` не спрацювало
    #    Використовуємо точне повідомлення про помилку з PDF
    print("Помилка. У списку присутнє нечислове значення")
```
